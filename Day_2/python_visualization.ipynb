```python
# Selects rows from Alabama, the second arguments are the columns
data_for_al = health_data.loc[health_data['State'] == 'AL', ['Population_Density','Poverty']]

#presents the top of this list
data_for_al.head()
```

```python
# Creating a nested list for these values

lst = [tuple(x) for x in data_for_al.values]
# every row is turned into a tuple from the ALABAMA data
#lst
```

```python
# CREATING A SCATTER PLOT FROM THIS DATA

scatter_plot = pygal.XY(stroke=False, x_title = 'Population Density', y_title = 'Poverty')
# XY is the scatter plot function, Stroke= False is a argument of the function
# Adding label to x-axis and y-axis

scatter_plot.title = 'Population Density vs Poverty Correlation'
# Adding a Title to the graph

scatter_plot.add('Alabama:Poverty',lst)
# 'AL' is the title of the points, second seciton is the entries of the plot (puts the points on the plot)


SVG(scatter_plot.render())
# Displays the current scatter plot


```

```python
# CHALLENGE: Make a graph like the above one for another state (CA)

#Creating a variable for the data, choosing california
data_for_ca = health_data.loc[health_data['State'] == 'CA', ['Population_Density','Poverty']]

# every row is turned into a tuple from the CALIFORNIA data
ca_lst = [tuple(x) for x in data_for_ca.values]


# XY is the scatter plot function, Stroke= False is a argument of the function and Adding label to x-axis and y-axis
scatter_plot_ca = pygal.XY(stroke=False, x_title = 'Population Density', y_title = 'Poverty')

# Adding a Title to the graph
scatter_plot_ca.title = 'Population Density vs Poverty Correlation'

# 'CA' is the title of the points, second seciton is the entries of the plot (puts the points on the plot)
scatter_plot_ca.add('California:Poverty',ca_lst)


# Displays the current scatter plot
SVG(scatter_plot_ca.render())

```

```python
# Adding the values for the state of texas to the same graph

data_for_tx = health_data.loc[health_data['State'] == 'TX', ['Population_Density','Poverty']]
lst2 = [tuple(x) for x in data_for_tx.values]

scatter_plot = pygal.XY(stroke=False, x_title = 'Population Density', y_title = 'Poverty')

scatter_plot.title = 'Population Density vs Poverty Correlation'

scatter_plot.add('AL',lst)

# This is where the new points are added
scatter_plot.add('TX', lst2)


SVG(scatter_plot.render())

```

```python
# Changind the x-axis incraments

data_for_tx = health_data.loc[health_data['State'] == 'TX', ['Population_Density','Poverty']]
lst2 = [tuple(x) for x in data_for_tx.values]


# adjusting the x range (look at final entry in the function 'xrange')
# scatter_plot.x_labels = (i for i in range (0,2000,100))

scatter_plot = pygal.XY(stroke=False, x_title = 'Population Density', y_title = 'Poverty', xrange = (1,100))
scatter_plot.title = 'Population Density vs Poverty Correlation'

# changes are here!

scatter_plot.add('AL',lst)
scatter_plot.add('TX', lst2)


SVG(scatter_plot.render())

scatter_plot.render_to_file('C:/Users/Juan Cardenas/Desktop/graph.svg')
```

```python
#USING A USE DEFINED FUNCTION

al = health_data.loc[health_data['State'] == 'AL', ['Poverty']]
poverty_mean_al = al['Poverty'].mean()
poverty_mean_al
```

```python
def mean_of_column(state_name, column_name):
    state_df = health_data.loc[health_data['State'] == state_name, [column_name]]
    return state_df[column_name].mean()
mean_of_column('TX','Poverty')
```

```python
# gives all the unique values in the state column
health_data['State'].unique()
# gives what type of object the argument is
type(health_data['State'].unique())

#makes the input into a list
#lst(health_data['State'].unique())
```

```python
# vertical Bars
bar_chart = pygal.HorizontalBar(legend_at_bottom=True, legend_at_botton_columns=11)
#moving the 
bar_chart.tite = 'Mean Povert Across all states'
for state in list(health_data['State'].unique()):
    bar_chart.add(state, mean_of_column(state, 'Poverty'))
   
SVG(bar_chart.render())

# To make horizontal bars, make .Bar() into .HorizontalBar() and vice versa
```
